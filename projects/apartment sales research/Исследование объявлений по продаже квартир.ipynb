# Предобработка данных

## Замена типа данных и замена пропусков

Начнём с перевода типа данных в колонке "first_day_exposition" на формат времени:

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data.head()
```

С датами разобрались, теперь можно перевести числовые значения в некоторых столбцах в целые значения.<br>
В столбцах с площадями такое делать, думаю, не стоит, чтобы не исказить данные. А вот в столбцах с растояниями есть смысл заменить значения на целые. Ещё есть смысл перевести значения в целые в столбце с ценами, этажами и балконами.<br><br>
**Приступим:**<br>
Для наглядности соберём все нужные нам столбцы в одну таблицу:

```python
data[
    ['last_price', 'floors_total', 'balcony', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000',
     'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition']].head(10)
```

Тут возникает одна загвоздка: во всех этих данных есть пропущенные значения, а значит функция для замены типа данных не даст нам провести замену.<br>
Но есть у меня гипотеза, что система сбора данных создавала пропуски там где информация отсутствовала. Например, если в анкете не вписывали балконы, или где отсутствовали поблизости парки, система ставила пропуск.<br>
В этом случае мы можем заменить пропуски на 0. Исключением будут расстояния. Так как, если мы заменим эти пропуски на 0, то исказим данные по расстояним до ключевых точек. Значит колонки с расстояниями менять не будем.<br>
Ещё не получится заменить тип в колонке дней продажи, так как не понятно из-за чего пропуски. Возможно объявление ещё активно, а может быть оно было снято с продажи. Предлагаю пропуски в этой колонке тоже оставить.
Уберём ненужные колонки и оставим подходящие:

```python
data[
    ['last_price', 'floors_total', 'balcony', 'parks_around3000', 'ponds_around3000']].head(10)
```

В колонке с ценами пропусков нет, остальные значения заменим. <br>
Есть ещё одно но: в колонке "floors_total" мы не можем указать нулевое количество этажей, значит заменим эти значения медианой, а стальное нулём:

```python
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data['balcony'] = data['balcony'].fillna(0)
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)

data[
    ['last_price', 'floors_total', 'balcony', 'parks_around3000', 'ponds_around3000']].head(10)
```

Теперь переведём данные в целые числа:

```python
data['last_price'] = data['last_price'].astype('int')
data['floors_total'] = data['floors_total'].astype('int')
data['balcony'] = data['balcony'].astype('int')
data['parks_around3000'] = data['parks_around3000'].astype('int')
data['ponds_around3000'] = data['ponds_around3000'].astype('int')

data[
    ['last_price', 'floors_total', 'balcony', 'parks_around3000', 'ponds_around3000']].head(10)
```

Выглядит намного приятнее!

## Работа с пропусками

Посмотрим, где у нас ещё остались пропуски:

```python
data.isnull().sum()
```

Можно ещё заменить пропуски в колонке "is_apartment" на False, чтобы получить колонку с булевыми значениями:

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Что ещё можно сделать с пропусками:
- В колонке "days_exposition" предполагаю, что значения пропущены в тех местах, где объявление вероятно ещё активно, либо оно было снято с продажи. Как ранее уже писал, можно их оставить.
- В колонках с расстояниями от ключевых точек пропуски скорее всего там, где этих ключевых точек нет, но заменить их на ноль мы не можем, это сильно может исказить данные. Указывать там какие-то средние значения - тоже может оказать вредное влияние. Тоже считаю, что стоит оставить пропуски в этих колонках.
- В колонке населённого пункта, боюсь придётся удалять строки, так как восстановить названия уже вряд-ли получится. К счастью их совсем немного и вряд ли это окажет какое-то негативное влияние.
- В "kitchen_area", целесообразней будет взять медианное значение, чем ставить 0.
- Аналогично, думаю, будет поступить с колонками жилой площади и высоты потолков.

Выполним все необходимые действия:

```python
data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_area'].median())
data['living_area'] = data['living_area'].fillna(data['living_area'].median())
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

Произвёл замены на медианные значения, теперь удалим строки с пропущенными населёнными пунктами:

```python
data.dropna(subset = ['locality_name'], inplace = True)
```

Посмотрим теперь, что получилось:

```python
data.isnull().sum()
```

Пропуски остались всего в нескольких столбцах, основные столбцы все заполнены практически без потерь.<br>
Думаю, теперь можно двигаться дальше!

## Работа с повторами

В названиях населённых пунктов наверняка есть повторные названия.<br>
Возьмём первые 10 наиболее встречающихся и посмотрим, есть-ли среди них что-то повторяющееся:

```python
data['locality_name'].value_counts().head(10)
```

Взглянем например на посёлок Мурино:

```python
data.loc[data['locality_name'] == 'посёлок Мурино']['locality_name'].count()
```

```python
data.loc[data['locality_name'] == 'поселок Мурино']['locality_name'].count()
```

```python
data.loc[data['locality_name'] == 'Мурино']['locality_name'].count()
```

Как видно, в столбце есть повторы одного и того-же населённого пункта. Наверняка и с другими городами такая-же ситуация.<br>
Чтобы это исправить, можно воспользоваться регулярными выражениями, заменить буквы ё на е, и удалить все слова которые не начинаются с большой буквы. Таким образом, в колонке остануться только названия населённых пунктов.

```python
import re 
def save_cities_words(row):
    #меняем Ё на Е
    locality = row['locality_name'].replace('ё','е')
    #выкидываем все слова, которые начинаются не с большой буквы
    return re.sub(r'\b[а-я]+\s*', "", locality).strip()
data['locality_name'] = data.apply(save_cities_words, axis=1)
```

```python
data['locality_name'].value_counts().head(10)
```

Таким образом мы избавились от повторов и получили тот же Топ-10, но с полными данными по населённым пунктам. Эти данные могут пригодиться в дальнейшем.

## Этаж квартиры

Создадим функцию, которая будет из имеющихся данных выводить нам категорию этажа: первый, последний или другой:

```python
def floor_name(data):
    if data['floor'] == 1:
        return 'первый'
    elif (data['floor'] == data['floors_total']) & (data['floor'] != 1):
        return 'последний'
    else:
        return 'другой'
```

Теперь применим функцию для создания колонки с категориями этажей:

```python
data['floor_category'] = data.apply(floor_name, axis=1)
```

```python
data['floor_category'].value_counts()
```

Добавил колонку с категориями этажа. Теперь можно делать выборки по категориям этажей.

Теперь посчитаем и создадим колонки с соотношением жилой и общей площади, а так-же соотношением площади кухни к общей:

```python
data['living_ratio_percent'] = (data['living_area'] / data['total_area'] * 100).round().astype('int')
data['kitchen_ratio_percent'] = (data['kitchen_area'] / data['total_area'] * 100).round().astype('int')

data[['total_area', 'living_area', 'kitchen_area',
      'living_ratio_percent', 'kitchen_ratio_percent']].head()
```
